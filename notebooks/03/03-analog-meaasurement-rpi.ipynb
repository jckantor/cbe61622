Verify operation

```micropython
import machine
import time

led = machine.Pin(25, machine.Pin.OUT)

def toggle_led(timer):
    led.toggle()

# run event loop
timer = machine.Timer(freq=20, mode=machine.Timer.PERIODIC, callback=toggle_led)

# timeout after 10 seconds
time.sleep(10)
timer.deinit()

# turn off led
led.value(0)
```

Measure Grove Light Sensor v1.2

```micropython
import machine
import time

adc = machine.ADC(0)
v = []

def print_adc(timer):
    val = adc.read_u16()
    v.append(val)
    
timer = machine.Timer(freq=200, mode=machine.Timer.PERIODIC, callback=print_adc)
time.sleep(1)
timer.deinit()

print(len(v))
```

```micropython
import utime
import machine
import _thread

# Connect the Grove LED to digital port
led = machine.Pin(18, machine.Pin.OUT)
    
stop_heartbeat = False
def heartbeat():
    global stop_heart_beat
    while True:
        led.toggle()
        utime.sleep(0.5)
        if stop_heartbeat:
            break         
    
background_thread = _thread.start_new_thread(heartbeat, ())


# measure signal
v = []
adc = machine.ADC(0)
k = 1000
while k > 0:
    val = adc.read_u16()
    v.append(val)
    utime.sleep(0.005)
    k -= 1

stop_heartbeat = True
```
